## Color Mapping
* So far we've only worked with RGB color spaces, in RGB codng, colors are modeled as a combination of Red, Green and Blue.
* In the 1970s HSL(hue, saturation, lightness) and HSV(hue, saturation, value) were developed as the alternative color models.
* HSL and HSV are more closely alinged with the way human vision actually perceives color.
* While in the this section we will deal almost exclusively with RGB images, its good idea to understand how to convert to HSL and HSV colorspaces.

```python
# Import all the tools 
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import the Image
img=cv2.imread(r"00-puppy.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### HLS (Hue,Lightness, Saturation)

```python
img=cv2.imread(r"00-puppy.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2HLS)
plt.imshow(img);
```

### HSV(Hue, Saturation, Value)

```python
img=cv2.imread(r"00-puppy.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(img);
```

```python
img=cv2.imread(r"00-puppy.jpg",0)
plt.imshow(img,cmap='gray')
```

## Morphological Operators
Morphological operators are sets of kernels that can achieve a variety of effects, such as reducing noise.

Certain operators are very good at reducing black points on a white background (and vice-versa)

Certain operators can also achieve an erosion and dilation effect that can add or erode from an existing image.

This effect is most easily seen on text data, so we will practice various morphological operators on some simple white text on a black background.

```python
# Import tools 
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img=np.zeros((600,800))
    font=cv2.FONT_HERSHEY_SIMPLEX
    black_img=cv2.putText(blank_img,text="ARYAN",org=(100,350),fontFace=font,fontScale=6,color=(255,255,255),thickness=30,lineType=cv2.LINE_AA)
    
    return black_img
```

```python
def show_img(img):
    fig=plt.figure(figsize=(10,15))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img=load_img()
show_img(img)
```

## Erosion
Erodes away boundaries of foreground objects. Works best when foreground is light color (preferrably white) and background is dark.

```python
kernel=np.ones((5,5),dtype=np.uint8)
```

```python
kernel
```

```python
result=cv2.erode(img,kernel,iterations=1)
show_img(result)
```

```python
img=load_img()
result1=cv2.erode(img,kernel,iterations=4)
show_img(result1)
```

### Opening 
Opening is erosion followed by dilation. Useful in removing background noise

```python
img=load_img()
show_img(img)
```

```python
# add background noise in the image
white_bg_noise=np.random.randint(low=0,high=2,size=(600,800))
white_bg_noise
```

```python
white_noise=white_bg_noise*255
noise_img=img+white_noise
```

```python
show_img(noise_img)
```

### Morphological operation (opening)

```python
opening=cv2.morphologyEx(noise_img,cv2.MORPH_OPEN,kernel)
show_img(opening)
```

### Closing
Useful in removing noise from foreground objects, such as black dots on top of the white text.

```python
img=load_img()
show_img(img)
```

```python
black_noise=np.random.randint(low=0,high=2,size=(600,800))
fg_noise=black_noise*-255
noise_img=img + fg_noise
noise_img[noise_img==-255]=0
show_img(noise_img)
```

```python
# Morphological operation
closing=cv2.morphologyEx(noise_img,cv2.MORPH_CLOSE,kernel)
show_img(closing)
```

### Morphological Gradient
Difference between dilation and erosion of an image

```python
img=load_img()
show_img(img)
```

```python
grad=cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
show_img(grad)
```

## Blending and pasting Images
* Blending images is done through the addWeighted function that uses both images and combine them.
* To blend images we use a simple formula new_pixel=alpha*pixel_1 + beta*pixel_2 + gamma
* For some computer vision systems, we'll want to be able to post our own image on the top of an already existing image or a video. We also want to have a "highlight" effect instead of just a solid box or empty rectangle.

* These are reffered to as Arithmetic Operations because OpenCV is simply performing simple common math with pixelsfor the final effect

### Overlaying Images of different sizes
We can use this quick trick to quickly overlap different sized images, simply reassigning the larger image's value to match the smaller image

```python
# Load the two image
plt.imshow(img1)
```

```python
large_img=img1.copy()
re_img=cv2.resize(img2,(500,500))
plt.imshow(re_img)
```

```python
img1.shape,re_img.shape
```

```python
# We will use simple pixel indexing to overlap small image onto the larger image
x_offset=0
y_offset=0
large_img[y_offset:y_offset+re_img.shape[0] , x_offset:x_offset + re_img.shape[1]]=re_img
```

```python
plt.imshow(large_img)
```

```python
# Let's try again and overlap small image on other position
large_img=img1.copy()
small_img=re_img.copy()
large_img[901:1401, 0:500]=small_img
plt.imshow(large_img)
```

### Blending together images of different sizes

Now, we learn how to blend images of different sizes. We need to select an ROI (Region Of Interest) in the first image, replace the values in that particular region with the values of the second image and then add them back in. However, the white background of the second image is quite annoying and our goal is to make only the logo visible. To do this we need to perform several masking procedures and bitwise operations.

```python
# Import Images
large_img=img1.copy()
plt.imshow(large_img)
```

```python
small_img=cv2.resize(img2,(500,500))
plt.imshow(small_img)
```

```python
rows,cols,chanels=small_img.shape
y_offset=1401-500
x_offset=934-500
```

```python
# Region of Interest
roi=large_img[901:1401,434:934]
```

```python
roi.shape
```

```python
plt.imshow(roi)
```

```python
# Convert image 2 to gray scale
small_gray=cv2.cvtColor(small_img,cv2.COLOR_RGB2GRAY)
plt.imshow(small_gray,cmap='gray')
```

```python
# Inversing the colors 
mask_inv=cv2.bitwise_not(small_gray)
```

It works ... bitwise. Assume the value of a pixel is 20010=110010002; then the bitwise not of that simply is 00110111b=55. Do that for all pixels.


```python
plt.imshow(mask_inv,cmap='gray')
```

```python
# create a white background
import numpy as np
white_background=np.full(small_img.shape,255,dtype=np.uint8)
```

```python
white_background.shape
```

```python
plt.imshow(white_background)
```

```python
bg=cv2.bitwise_or(white_background,white_background,mask=mask_inv)
plt.imshow(bg)
```

```python
fg=cv2.bitwise_or(small_img,small_img,mask=mask_inv)
plt.imshow(fg)
```

```python
final_roi=cv2.bitwise_or(roi,fg)
plt.imshow(final_roi)
```

```python
large_img[901:1401,434:934]=final_roi
```

```python
plt.imshow(large_img)
```

```python
fig=plt.figure(figsize=(10,15))
ax=fig.add_subplot(111)
ax.imshow(large_img)
```

```python
final_roi2=cv2.bitwise_or(roi,bg)
plt.imshow(final_roi2)
```

```python
large_img[901:1401,434:934]=final_roi2
```

```python
plt.imshow(large_img)
```

```python
fig=plt.figure(figsize=(10,15))
ax=fig.add_subplot(111)
ax.imshow(large_img)
```

## Practicing (Image Processing)

```python
#Import all the needed tools
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig=plt.figure(figsize=(10,15))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
# Import giraffes image
giraffe=cv2.imread('giraffes.jpg')
giraffe_rgb=cv2.cvtColor(giraffe,cv2.COLOR_BGR2RGB)
display_img(giraffe_rgb)
```

### Applying binary threshold on to the image

```python
giraffe_gray=cv2.imread('giraffes.jpg',0)
ret,thres=cv2.threshold(giraffe_gray,120,255,cv2.THRESH_BINARY)
display_img(thres,cmap='gray')
```

### Creating a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)

```python
kernel=np.ones((4,4),np.float32)/10
kernel
```

```python
blur_img=cv2.filter2D(giraffe,-1,kernel)
blur_img=cv2.cvtColor(blur_img,cv2.COLOR_BGR2RGB)
display_img(blur_img)
```

### Creating a Horizontal Sobel Filter (sobelx) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.

```python
img=giraffe_gray
display_img(img,cmap='gray')
```

```python
img_x=cv2.Sobel(img,cv2.CV_64F,1,0,5)
display_img(img_x,cmap='gray')
```

### Ploting the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image.

```python
img=giraffe
```

```python
color=['b','g','r']
for i,col in enumerate(color):
    hist=cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color=col)
plt.title("Histogram for the b,g,r color channels of the giraffe image")
plt.show();
```

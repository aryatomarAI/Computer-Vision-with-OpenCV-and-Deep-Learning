## Opening Images files using OpenCV
**What is OpenCV ?**

OpenCV(Open Source Computer Vision) is a library of programming functions mainly aimed at real-time computer vision. OpenCv is created by Intel in 1999,it is written in C++.

It contains many popular algorithms for computer Vision, including object detection and tracking algorithms built-in.




`In this notebook we will learn how to open image files with OpenCV in both a notebook and a python script.`
    

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img=cv2.imread(r"C:\Users\DELL\Desktop\Computer-Vision-with-OpenCV-and-Deep-Learning\Numpy\00-puppy.jpg")
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)  # here our image colors are kind of weird because in OpenCV color channels are in order BGR, whereas in matplotlib it's RGB
```

**MATPLOTLIB-RGB**

**OpenCV-BGR**

```python
# we can fix this by with help of OpenCV
fix_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
# Now look at that we got our original image
```

### Converting our image to grayscale

```python
img_gray=cv2.imread(r"C:\Users\DELL\Desktop\Computer-Vision-with-OpenCV-and-Deep-Learning\Numpy\00-puppy.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray');
```

### Resizing the image

```python
fix_img.shape
```

```python
resize_img=cv2.resize(fix_img,(1200,1000))   # src,size of the image
plt.imshow(resize_img)
```

```python
# Resizing our image with ratios
new_img=cv2.resize(fix_img,(0,0),fix_img,0.5,0.5) # width ratio, height ratio
plt.imshow(new_img);
```

```python
new_img.shape   # Our image size i.e pixels becomes half
```

### Flipping our image

we will flip method of OpenCv to flip our images

```python
new_img1=cv2.flip(fix_img,0)  # 0 for downwards
plt.imshow(new_img1);
```

```python
new_img1=cv2.flip(fix_img,1)  # 1 for left shift
plt.imshow(new_img1);
```

```python
new_img1=cv2.flip(fix_img,-1)  # Downward left shift
plt.imshow(new_img1);
```

### Save an Image file

```python
cv2.imwrite("resized_image.jpg",new_img)
```

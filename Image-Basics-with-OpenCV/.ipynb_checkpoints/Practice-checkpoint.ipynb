## Practicing (Image basics with OpenCV)

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read the image
my_dog=cv2.imread(r"dog_backpack.jpg")
my_dog
```

```python
plt.imshow(my_dog)
```

```python
fix_dog_img=cv2.cvtColor(my_dog,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_dog_img)
```

### Flip the image upside down

```python
flip_img=cv2.flip(fix_dog_img,flipCode=0)
```

```python
plt.imshow(flip_img)
```

### Drawing an empty RED rectangle around the dogs face and displaying the image 

```python
img=fix_dog_img.copy()
plt.imshow(img)
```

```python
cv2.rectangle(img,(175,350),(650,750),(250,0,0),5)
```

```python
plt.imshow(img)
```

**Drawing a BLUE TRIANGLE in the middle of the image. The size and angle is up to us, but it should be a triangle (three sides) in any orientation.**

```python
vertices= np.array([[300,750],[450,400],[600,750]],dtype=np.int32)
```

```python
pts=vertices.reshape((-1,1,2))
```

```python
pts.shape
```

```python
my_img=fix_dog_img.copy()
```

```python
cv2.polylines(my_img,[pts],isClosed=True,color=(0,0,250),thickness=10)
```

```python
plt.imshow(my_img)
```

### Fill the triangle

```python

cv2.fillPoly(my_img,[pts],color=(0,0,255))
plt.imshow(my_img)
```

**Create a script that opens the picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.**

```python
import numpy
import cv2

def draw_circles(event,x,y,flags,param):
    if event==cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),50,(0,0,255),10)

img=my_dog
cv2.namedWindow(winname='dog_window')
cv2.setMouseCallback('dog_window',draw_circles)

while True:
    cv2.imshow("dog_window",img)
    if cv2.waitKey(5) & 0xFF==27:
        break
        
cv2.destroyAllWindows()
```

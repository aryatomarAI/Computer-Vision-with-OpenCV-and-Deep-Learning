## Canny Edge Detection
* Canny Edge Detector one of the most popular edge detection algorithm.
* It was developed in 1986 by John Canny and is multi-stage algorithm.

**Steps of Canny**

1. It apply Gaussian filter to smooth the image in order to remove the noise.
2. Find the intensity gradient of the image.
3. It apply non-maximum suppression to get rid of spurious response to edge detection.
4. Apply double threshold to determine potential edges.
5. Track edge by hystersis: Finalize the detection of edges by suppressing all the other edges that are weak and not connected to strong edges.

**Note:** For high resolution images where you only want general edges, it is usually a good idea to apply a custom blur before applying the canny algorithm

* The Canny Algorithm also requires a user to decide on low and high threshold values.

```python
# Import all the required modules
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load image
dog_face=cv2.imread("sammy_face.jpg")
# We don't have to convert color channels in case of edge detection
```

```python
# Canny Edge detection
edges=cv2.Canny(image=dog_face,threshold1=250,threshold2=180)
```

```python
plt.imshow(edges)
```

### Choosing Thresholds

```python
# median pixel value
med_val=np.median(dog_face)
```

```python
# Lower bound is either 0 or 70% of the median value, whichever is higher
lower=int(max(0,0.7*med_val))
```

```python
# Upper bound is either 255 or 30% above the median, whichever is lower
upper=int(min(255,1.3*med_val))
```

```python
edges=cv2.Canny(dog_face,lower,upper)
plt.imshow(edges)
```

```python
# It's a high resolution image we have to blur it to decrease noise from the image
blur_img=cv2.blur(src=dog_face,ksize=(5,5))
```

```python
edges=cv2.Canny(blur_img,lower,upper)
plt.imshow(edges)
```

```python
edges=cv2.Canny(blur_img,lower,upper+50)
plt.imshow(edges);
```

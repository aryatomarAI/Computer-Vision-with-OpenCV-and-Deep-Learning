## Russian Plate Bluring

We have to detect the number plate of the vehicle, draw a rectangle around it and then blur it 

Step 1. Load the image 

Step 2. Display the image on large scale

Step 3. Turn the image into gray Scale

Step 4. load the HaarCascade plate detection file

Step 5. create a function to detect and draw rectangle around the number plate.

Step 6. Create a function that detect and blur the number plate in the image and return it

```python
# Import all the needed modules
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the image
vehicle_img=cv2.imread('car_plate.jpg')
vehicle_img=cv2.cvtColor(vehicle_img,cv2.COLOR_BGR2RGB)
grey_plate=cv2.cvtColor(vehicle_img,cv2.COLOR_RGB2GRAY)
```

```python
# Display function
def display(img,cmap=None):
    fig=plt.figure(figsize=(10,12))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
display(vehicle_img)
```

```python
display(grey_plate,cmap='gray')
```

```python
#load plate cascade file
plate_cascade=cv2.CascadeClassifier('haarcascades/haarcascade_russian_plate_number.xml')
```

```python
# Create a function to detect the russian car plate
def detect_plate(img):
    car_plate=img.copy()
    color_img=vehicle_img.copy()
    
    plate_rects=plate_cascade.detectMultiScale(car_plate,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in plate_rects:
        cv2.rectangle(color_img,(x,y),(x+w,y+h),(255,0,0),3)
    
    return color_img
```

```python
plate_detect=detect_plate(grey_plate)
display(plate_detect)
```

```python
# Now let's bind the above code into a function

def detect_and_blur(img):
    car_img=img.copy()
    gray_img=cv2.cvtColor(car_img,cv2.COLOR_RGB2GRAY)
    
    car_plate_rects=plate_cascade.detectMultiScale(gray_img,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in car_plate_rects:
        number_plate=car_img[y:y+h,x:x+w]
    
    blur_plate=cv2.medianBlur(number_plate,7)
    
    car_img[y:y+h,x:x+w] = blur_plate
    
    return car_img
    
    
```

```python
blured_plate=detect_and_blur(vehicle_img)
display(blured_plate)
```

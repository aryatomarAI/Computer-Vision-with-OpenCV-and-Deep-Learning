## Object Detection with OpenCV
Understanding a variety of object detection methods. Let's start with Template Matching :

**Terminal Matching:**
    
    Simply looking for an exact copy of an image in another image.
    
    Template Matching is the simplest form of object detection.
    
    It simply scans a large image for a provided template by sliding the template target image across the larger image.
    
    The main option that can be adjusted is the comparison method used as the target template is slid across the larger image.
    
    The methods are all some sort of corelation based metric.





```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import the image

img_sammy=cv2.imread('sammy.jpg')
img_sammy=cv2.cvtColor(img_sammy,cv2.COLOR_BGR2RGB)
plt.imshow(img_sammy)
```

```python
# import sammy's face

sammy_face=cv2.imread("sammy_face.jpg")
sammy_face=cv2.cvtColor(sammy_face,cv2.COLOR_BGR2RGB)
plt.imshow(sammy_face)
```

### Template Matching Methods
We are using 6 Template Matching modes and all the 6 methods are in a list for comparison

```python
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

To execute these methods we have to use eval function 

```python
height,width,channels=sammy_face.shape
```

```python
for m in methods:
    # Copy of the image
    full_copy=img_sammy.copy()
    
    # method
    method=eval(m)
    
    # Template Matching
    
    res=cv2.matchTemplate(full_copy,sammy_face,method)
    
    min_value,max_value,min_loc,max_loc=cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left=min_loc
    else:
        top_left=max_loc
        
    height, width, channels=sammy_face.shape
    
    bottom_right=(top_left[0]+width,top_left[1]+height)
    
    # drawing a rectangle around the dog face
    cv2.rectangle(full_copy,top_left,bottom_right,(255,0,0),10)
    
    # plot heat map images and show images with rectangle
    plt.subplot(121)
    plt.imshow(res)
    plt.title("Heat map of the template matching")
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title("Detecting of template")
    plt.suptitle(m,y=1.1)
    
    plt.tight_layout(pad=1.0)
    plt.show()
    
    
    print('\n')
    print('\n')
    
    
    
```
